# Topic modelling con Scikit-learn

Effettuiamo il topic modelling e secondariamente "classifichiamo le frasi in aparrtenenti ad un determinato topic e non appartenenti". Cosi facendo creiamo una nuova lista dove le fras inserite rappresentano una sintesi estrattiva del libro

```python
news_file = open('immensola.txt', 'r', encoding='utf-8')
news = news_file.read().splitlines()

print(news)
```

## Preprocessiamo i dati

```python
len(news)
```

```python
#rimuoviamo le stringhe vuote
while("" in news):
    news.remove("") 
```

```python
len(news)
```

```python
#rimuoviamo "-" dalle stringhe
news_2=[]
for i in news:
    i=i.replace("- ", "")
    news_2.append(i)
news=news_2
```

```python
#len(news)
news[0]
```

```python
import re
news_2=[]
for y, s in enumerate(news):
    print(y)
    s = re.sub(r'[^\w\s]',' ',s)
    s= s.strip() #returns a new string after removing any leading and trailing whitespaces including tabs (\t).

    news_2.append(s)
    
news=news_2
```

## LDA con Bag of Words

```python
#trasformiamo tutto in minuscolo
news_2=[]

for i in news:
    i=i.lower()
    news_2.append(i)
    
news=news_2
len(news)
```

```python
#rimuoviamo le stringhe vuote (per la 2a volta!)
while("" in news):
    news.remove("") 
```

```python
#rimuoviamo le stop_words
from nltk.tokenize import word_tokenize

import spacy
sp = spacy.load('it_core_news_sm')

all_stopwords = sp.Defaults.stop_words
all_stopwords.add('disse') #aggiungo delle stopwords a mia scelta
all_stopwords.add('rispose')
all_stopwords.add('d') #rimuoviamo gli articoli che restano soli dopo aver tolto l'apostrofo
all_stopwords.add('l')
all_stopwords.add('l')

news_2=[]

for text in news:
    text_tokens = word_tokenize(text)
    tokens_without_sw= [word for word in text_tokens if not word in all_stopwords]
    i = ' '.join(tokens_without_sw)
    news_2.append(i)

news=news_2
```

```python
news
```

```python
#rimuoviamo le stringhe vuote (per la 3a volta!)
while("" in news):
    news.remove("") 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
news_split=news
bow = CountVectorizer(max_features=5000, lowercase=True)
X = bow.fit_transform(news_split)
X.shape #questa è un matrice sparsa
```

Adesso creiamo il modello LDA usando la classe *LatentDirichletAllocation* di sklearn, il mumero di topic va definito a priori all'interno del parametro n_components, impostiamolo a 25.

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA

n_topics = 5

lda = LDA(n_components=n_topics, max_iter=100, verbose=True)
lda.fit_transform(X)
```

Il modello è pronto ! Cerchiamo di farci un'idea di cosa i nostri topic rappresentano visualizzando le 10 parole più comuni per ognuno.

```python
n_words = 20

for index, topic in enumerate(lda.components_):
  print("\nTOPIC %d - %d parole più popolari" % (index, n_words))
  print([bow.get_feature_names()[i] for i in topic.argsort()[-n_words:]])
```

```python
#EXTRA!!!!!!!!!!!!!!!!!!!!!!!!!! -> visualizziamo graficamente i topic
import pyLDAvis.sklearn
lda_viz= pyLDAvis.sklearn.prepare(lda, X, bow)
pyLDAvis.display(lda_viz)
```

# Topic modelling usando il TF-IDF

mentre nella sezione precedente abbiamo creato un riassuntore usando il topic modelling ed il bag of word (abbreviato bow) in questa situazione invece usiamo il modello tf-idf, poi eventualemente valutiamo se i topic sono sovrapponibili

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words="english")
X = tfidf.fit_transform(news_split)
X.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA

n_topics = 5

lda = LDA(n_components=n_topics, max_iter=100, verbose=True)
lda.fit_transform(X)
```

```python
n_words = 10

for index, topic in enumerate(lda.components_):
  print("\nTOPIC %d - %d parole più popolari" % (index+1, n_words))
  print([bow.get_feature_names()[i] for i in topic.argsort()[-n_words:]])
```

```python
import pyLDAvis.sklearn

lda_viz = pyLDAvis.sklearn.prepare(lda, X, tfidf, mds='tsne')
pyLDAvis.display(lda_viz)
```
